```python
import autogen 
import config
```

```python
config_list = [
    
        {
        "model": "gemini-pro",
        "api_key": "",
        "api_type": "google"
    }
    
]
```

```python
def main():
   
    assistant = autogen.AssistantAgent(
        name="Assistant",
        llm_config={
            "config_list":config_list
        }
    )
    user_proxy = autogen.UserProxyAgent(
        name="User",
        human_input_mode="TERMINATE",
        code_execution_config = {
                "work_dir":"coding",
                "use_docker":False,
            }
        
    )


    user_proxy.initiate_chat(assistant,message="PLot  a chart of META and TESLA stock price")
```

```python
main()
```
